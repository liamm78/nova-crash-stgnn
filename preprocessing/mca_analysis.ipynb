```python
import pandas as pd
import prince

# Note: Try on Virginia. 
db = pd.read_csv("loudoun_county.csv")
pd.set_option('display.max_columns', None)

# db_columns = ["Light Condition"]
# db = db[db_columns]
db.head(40)
```

```python
# We wish to see if Light Condition relates to Crash Severity. We do a Chi-squared goodness of fit

from scipy.stats import chi2_contingency


contingency = pd.crosstab(db['Light Condition'], db['Crash Severity'])
print(contingency)
chi2, p, dof, expected = chi2_contingency(contingency)
print("Chi-squared:", chi2, "p-value:", p)


residuals = (contingency - expected) / (expected ** 0.5)

# Sum absolute residuals per Light Condition to rank importance
importance = residuals.abs().sum(axis=1).sort_values(ascending=False)
print("Ranking of Light Conditions by importance:")
importance_percent = 100 * importance / importance.sum()
print(importance_percent)

print("types")
print(db['Light Condition'].value_counts(normalize=True))
```

```python
db = db[['Roadway Alignment', 'Roadway Surface Type', 'Roadway Surface Condition','Roadway Description']]

mca = prince.MCA(
    n_components=6,
    n_iter=100,
    check_input=True,
    engine='sklearn',
)
mca = mca.fit(db)
```

```python
one_hot = pd.get_dummies(db)

print(mca.eigenvalues_summary)

mca.column_contributions_.head(50).style.format('{:.0%}')
```
